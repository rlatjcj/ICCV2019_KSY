```python
import os
import os.path as path
```

```python
data_base_path = 'metadata'
```

```python
metadata_path = path.join(data_base_path, 'Metadata')
label_filename = path.join(metadata_path, 'class-descriptions-boxable.csv')

with open(label_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
    
labels = [line.split(',') for line in lines]
label_dict = {label_name:real_name for label_name, real_name in labels}
```

```python
label_dict
```

```python
box_path = path.join(data_base_path, 'Boxes')
train_box_filename = path.join(box_path, 'train-annotations-bbox.csv')
val_box_filename = path.join(box_path, 'validation-annotations-bbox.csv')
test_box_filename = path.join(box_path, 'test-annotations-bbox.csv')
```

```python
with open(train_box_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
labels = [line.split(',')[2] for line in lines[1:]]
train_counted_labels = {label:0 for label in label_dict.keys()}
for label in labels:
    train_counted_labels[label] += 1
    
with open(val_box_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
labels = [line.split(',')[2] for line in lines[1:]]
val_counted_labels = {label:0 for label in label_dict.keys()}
for label in labels:
    val_counted_labels[label] += 1
    
with open(test_box_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
labels = [line.split(',')[2] for line in lines[1:]]
test_counted_labels = {label:0 for label in label_dict.keys()}
for label in labels:
    test_counted_labels[label] += 1
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import imagesize
from tqdm import tqdm_notebook as tqdm
```

---

```python
analyze_base_path = 'analyze'
for_retina_path = path.join(analyze_base_path, 'for_retina')
os.makedirs(for_retina_path, exist_ok=True)
```

```python
with open(train_box_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
informs = [line.split(',') for line in lines[1:]]
informs = [[info[0],*info[4:8],info[2]] for info in informs]
informs[:3]
```

```python
with open(train_box_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
informs = [line.split(',') for line in lines[1:]]
informs = [[info[0],*info[4:8],info[2]] for info in informs]

image_path = 'train'
temp_image_name = ''

train_for_retina_name = path.join(for_retina_path, 'train_for_retina.csv')
with open(train_for_retina_name, 'w', encoding='utf-8') as f:
    for img_name,  x1, x2, y1, y2, class_label in tqdm(informs):
        img_fullpath = path.join(image_path, img_name+'.jpg')
        if img_name != temp_image_name:
            img_x, img_y = imagesize.get(img_fullpath)
            temp_image_name = img_name
        
        x1 = img_x * float(x1)
        y1 = img_y * float(y1)
        x2 = img_x * float(x2)
        y2 = img_y * float(y2)
        
        x1 = int(x1)
        y1 = int(y1)
        x2 = int(x2)
        y2 = int(y2)

        f.write(','.join([img_fullpath.replace(path.sep,'/'), 
                              str(x1), str(y2), str(x2), str(y1), 
                              label_dict[class_label]])+'\n')
```

```python
with open(val_box_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
informs = [line.split(',') for line in lines[1:]]
informs = [[info[0],*info[4:8],info[2]] for info in informs]

image_path = 'validation'
temp_image_name = ''

val_for_retina_name = path.join(for_retina_path, 'validation_for_retina.csv')
with open(val_for_retina_name, 'w', encoding='utf-8') as f:
    for img_name, x1, x2, y1, y2, class_label in tqdm(informs):
        img_fullpath = path.join(image_path, img_name+'.jpg')
        if img_name != temp_image_name:
            img_x, img_y = imagesize.get(img_fullpath)
            temp_image_name = img_name
        
        x1 = img_x * float(x1)
        y1 = img_y * float(y1)
        x2 = img_x * float(x2)
        y2 = img_y * float(y2)
        
        x1 = int(x1)
        y1 = int(y1)
        x2 = int(x2)
        y2 = int(y2)

        f.write(','.join([img_fullpath.replace(path.sep,'/'),
                              str(x1), str(y2), str(x2), str(y1), 
                              label_dict[class_label]])+'\n')
```

```python
with open(test_box_filename, 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()
informs = [line.split(',') for line in lines[1:]]
informs = [[info[0],*info[4:8],info[2]] for info in informs]

image_path = 'test'
temp_image_name = ''

test_for_retina_name = path.join(for_retina_path, 'test_for_retina.csv')
with open(test_for_retina_name, 'w', encoding='utf-8') as f:
    for img_name,  x1, x2, y1, y2, class_label in tqdm(informs):
        img_fullpath = path.join(image_path, img_name+'.jpg')
        if img_name != temp_image_name:
            img_x, img_y = imagesize.get(img_fullpath)
            temp_image_name = img_name
        
        x1 = img_x * float(x1)
        y1 = img_y * float(y1)
        x2 = img_x * float(x2)
        y2 = img_y * float(y2)
        
        x1 = int(x1)
        y1 = int(y1)
        x2 = int(x2)
        y2 = int(y2)

        f.write(','.join([img_fullpath.replace(path.sep,'/'),
                              str(x1), str(y2), str(x2), str(y1), 
                              label_dict[class_label]])+'\n')
```

```python
analyze_base_path = 'analyze'
frequency_path = path.join(analyze_base_path, 'frequency')
os.makedirs(frequency_path, exist_ok=True)
```

```python
temp_labels = sorted([inform for inform in train_counted_labels.items()], 
                         key=lambda inform: inform[1], reverse=True)

real_labels = [label_dict[label_inform[0]] for label_inform in temp_labels]
label_cnts = [label_inform[1] for label_inform in temp_labels]

plt.figure(figsize=(8,4), dpi=100)
plt.title('Train Data Frequency')
plt.bar(real_labels, label_cnts, width=1)
plt.xticks([])
plt.savefig(path.join(frequency_path, 'Train Data Frequency.png'))
plt.show()

cnt_filename = path.join(frequency_path, 'train_data_frequency.csv')
with open(cnt_filename, 'w', encoding='utf-8') as f:
    f.write(','.join(['class_label', 'real_label', 'count'])+'\n')
    for class_label, cnt in temp_labels:
        f.write(','.join([class_label, label_dict[class_label], str(cnt)])+'\n')
```

```python
percentages = np.zeros_like(label_cnts, dtype=np.float32)
cnt_sum = np.sum(label_cnts)
sum_temp = cnt_sum

for i, cnt in enumerate(label_cnts):
    percentages[i] = sum_temp / cnt_sum
    sum_temp -= cnt

plt.figure(figsize=(8,4), dpi=100)
plt.title('Train Data Percentage')
plt.bar(np.arange(len(percentages)), percentages, width=1)
plt.axhline(y=0.1, color='r', linestyle='--', label='0.9')
plt.axhline(y=0.05, color='y', linestyle='--', label='0.95')
plt.axhline(y=0.01, color='g', linestyle='--', label='0.99')
# plt.xticks([])
plt.legend()
plt.savefig(path.join(frequency_path, 'train_data_percentage.png'))
plt.show()
```
